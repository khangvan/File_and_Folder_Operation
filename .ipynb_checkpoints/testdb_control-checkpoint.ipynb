```python
def lines_that_equal(line_to_match, fp):
    return [line for line in fp if line == line_to_match]

def lines_that_contain(string, fp): #common use
    return [line for line in fp if string.lower()+"=".replace(" ","") in line.lower().replace(" ","")] # lowwer + add =

def lines_that_start_with(string, fp):
    return [line for line in fp if line.lower().replace(" ","").startswith(string.lower()+"=".replace(" ",""))]

def lines_that_end_with(string, fp):
    return [line for line in fp if line.endswith(string)]
def generate_lines_that_equal(string, fp):
    for line in fp:
        if line == string:
            yield line
with open(filename, "r") as fp:
    for line in lines_that_start_with("Name", fp):
        print (line)
```

```python
from tempfile import mkstemp
from shutil import move
from os import fdopen, remove

def replace(file_path, pattern, subst):
    #Create temp file
    fh, abs_path = mkstemp()
    with fdopen(fh,'w') as new_file:
        with open(file_path) as old_file:
            for line in old_file:
                new_file.write(line.replace(pattern, subst))
    #Remove original file
    remove(file_path)
    #Move new file
    move(abs_path, file_path)
replace(filename,rf"Mode = "STD"","test")
```

```python
import os


class dltest():
    def __init__(self, filename):
        self._filename =filename
        print(filename)
    def run(self):
        df=self.converttodf(self._filename) 
        df2= self.shappingdata(df)
        self.df=df2
        return df2
    def reborn(self,path):
        self.regenerate(self.df,path)
    @staticmethod
    def converttodf(filename):
        mylist=[]
       
        if not os.path.isfile(filename):
            print('File does not exist.')
        else:
        # Open the file as f.
        # The function readlines() reads the file.
            with open(filename) as f:
                content = f.read().splitlines()

            # Show the file contents line by line.
            # We added the comma to print single newlines and not double newlines.
            # This is because the lines contain the newline character '\n'.
            for line in content:
        #         print(line.split(r"="))
                myvalue=line.split(r"=")
                if (myvalue[0]==""): # remove all row empty
                    pass
                else:
                    mylist.append(line.split(r"="))

        import pandas as pd
        df=pd.DataFrame(mylist,columns=["key","value"])
    #         df.head(1000)
        df['source']= filename
        return df

    @staticmethod
    def shappingdata(df):
        import numpy as np
        df['message']=df['key'].apply(lambda x: ";" if x[:1] ==";" else "")
        df['key']=df['key'].apply(lambda x: x.replace(";",""))     # remove to make it eaaisr reading
        
        df['value'] = df['value'].replace([None],np.nan )
        df['value'] = df['value'].replace(np.nan,"")
     
        
        df.head(100)
        df['module']=df['key'].apply(lambda x: x if x[:1] =="[" else np.nan)
        
        df["module"]=df["module"].ffill() # 
#         df["value"]=df.replace({'value': {"None": ""}}).ffill() # 
        
        df['key']=df['key'].apply(lambda x: "" if(x[:1]=="[") else x) 
        return df

    @staticmethod
    def regenerate(df, savefile):
#         savefile=fr"C:\idpmdc\demofile3.tdb"
        f = open(savefile, "w")
        for index, row in df.iterrows():
            data= '\r{}\r\n'.format(row['module']) if row['key']==""   else "{}{}{}\r".format((row['message']),row['key'] ,"="+row['value'] if len(row['value'])>0 else "")
        #     print(data)
            import numpy as np
            print(data)
            f.write( data)
        f.close()
        print("done generate file at {}".format(savefile))
        
        
filename=fr"C:\\idpmdc\\dltest.tdb"        
app=dltest(filename)
mydf=app.run()
app.reborn("C:\idpmdc\demofile3.tdb")
mydf
```
